```
import requests

from bs4 import BeautifulSoup
```

```
r = requests.get('https://www.elpais.com')

c = r.content
```

```
soup = BeautifulSoup(c, 'html.parser')
```

```
print(soup.prettify)
```

```
"""
Mini srcipt mètode per extreure tots els titulars de la pagina
"""
datos = soup.findAll('h2')

cosas = [dato.text.strip() for dato in datos]

for _ in cosas:
  print(_)
```

```
"""
Mini srcipt mètode per extreure tots els autors
"""
metadatos = soup.findAll('div',{'class':'articulo-metadatos'})

for _ in [m.text.strip() for m in metadatos]:
  print(_.replace('\n',''))
```

```
"""
Mini script extreure mini resumillo de les noticies mes peus de foto
"""
articles = soup.findAll('article')

for _ in [m.text.strip() for m in articles]:
  print(_.replace('\n',''))
  print('------------------------------------')
```

```
"""
Mini script extreure mini num comentaris
"""
coments = soup.findAll('a',{'class':'comentarios'})

for _ in [m.text.strip() for m in coments]:
  print(_)
  print('------------------------------------')
```

```
articles = soup.findAll('a')

for _ in articles:
  print(_)
  print('TEXT   |',_.text)
  print('STRIP  |',_.text.strip().replace('\n',''))
  print('------------------------------------')
```
